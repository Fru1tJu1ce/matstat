```python
import pandas as pd
import statsmodels.formula.api as smf


link = r'http://d396qusza40orc.cloudfront.net/statistics/lec_resources/states.csv'
df = pd.read_csv(link)
model = smf.ols(formula='poverty ~ metro_res + white + hs_grad + female_house', data=df)
result = model.fit()
new_frame = df[['metro_res', 'white', 'hs_grad', 'poverty', 'female_house']].copy()
```

```python
import pandas as pd
import numpy as np

corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
import matplotlib.pyplot as plt


fig, axes = plt.subplots(5,5)
fig.suptitle("Scatter plots", fontsize=30)
plt.rcParams['figure.figsize'] = [30, 30]


for frame_1 in enumerate(new_frame):
    for frame_2 in enumerate(new_frame):
        axes[frame_1[0], frame_2[0]].scatter(df[frame_2[1]], df[frame_1[1]])
        axes[frame_1[0], frame_2[0]].tick_params(labelsize=20)
        
        if frame_2[0] == 0:
            axes[frame_1[0], frame_2[0]].set_ylabel(frame_1[1], fontsize=25)
        if frame_1[0] == 4:
            axes[frame_1[0], frame_2[0]].set_xlabel(frame_2[1], fontsize=25)
            
        b1, b0 = np.polyfit(df[frame_2[1]], df[frame_1[1]], 1)
        axes[frame_1[0], frame_2[0]].plot(df[frame_2[1]], b0 + b1*df[frame_2[1]], color='red')

```
